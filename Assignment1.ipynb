Question 1: Sum of two numbers

```python
def sum_two(arr:list,target:int)->list:
  lst_ind=[]
  for i in range(len(arr)):
    for j in range(i+1,len(arr)):
      if (arr[i]+arr[j]==target):
        lst_ind.append(i)
        lst_ind.append(j)
        break
      else:
        continue
  return lst_ind

```

```python
sum_two([2,3,5,7,11],9)
```

Question 2: return the number of elements in nums which are not equal to val

```python
def num_ele(arr:list,val:int)->list:
  count=0
  new_arr=[]
  for i in range(len(arr)):
    if arr[i]!=val:
      count=count+1
      new_arr.append(arr[i])
    else:
      new_arr.append('_')

  return count,new_arr


```

```python
num_ele([22,2,2,3,3],3)
```

Question 3 : Index of val in arr

```python
def ind_arr(arr:list,val:int)->int:
  try:
    arr.sort()
    return arr.index(val)
  except ValueError:
    arr.append(val)
    arr.sort()
    return arr.index(val)
```

```python
ind_arr([2,4,5,6,7],7)
```

```python
ind_arr([2,4,5,6,7],3)
```

Question 4 : Plus one of digits

```python
def plus_one(arr:list)->int:
  arr=[str(i) for i in arr]
  strng=""
  strng = strng.join(arr)
  return int(strng)+1

```

```python
plus_one([1,2,3])
```

Question 5: merge two lists

```python
def merge_list(num1:list,num2:list,m:int,n:int):
  return num1[:m-1].extend(num2[:n-1])

```

```python
merge_list(num1 = [1,2,3,0,0,0],num2 = [2,5,6],m=3, n = 3)
```

Question 7 : Shift zeros

```python
def shift_zeros(arr:list):

  n = len(arr)
  j = 0
  for i in range(n):
      if arr[i] != 0:
          arr[j], arr[i] = arr[i], arr[j]  # Partitioning the array
          j=j+1
  return arr
```

```python
shift_zeros([0,1,0,3,12])
```

Question 8: correct and find the error in numbers

```python
from collections import Counter
def err_num(arr:list):
  c=Counter(arr)
  l=[0,0]
  for i in range(1,len(arr)+1):
    if c[i]==2:
      l[0]=i
    if c[i]==0:
      l[1]=i
  return l
```

```python
err_num([1,2,2,4])
```
